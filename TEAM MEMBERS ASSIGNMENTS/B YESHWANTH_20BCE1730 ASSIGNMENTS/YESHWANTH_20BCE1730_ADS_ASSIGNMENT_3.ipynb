## 7)Check for Categorical columns and perform encoding.

```python
from sklearn.preprocessing import LabelEncoder               #Label Encoding
```

```python
le = LabelEncoder()
```

```python
data.head()
```

```python
data.info()
```

```python
columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning']
for col in columns:
 data[col] = le.fit_transform(data[col])
```

```python
data.head()
```

```python
data = pd.get_dummies(data, columns=['furnishingstatus'])              #One hot encoding
```

## 8)Split the data into dependent and independent variables.

```python
y = data.loc[:, 'price':'price']                             #Dependent variables
y                                        
```

```python
x = data.drop(columns=['price'], axis=1)                      #Independent variables
x
```

## 9)Scale the independent variables

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
name = x.columns
x_scaled = scale.fit_transform(x)
```

```python
x_scaled
```

```python
x = pd.DataFrame(x_scaled, columns=name)
x
```

## 10)Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## 11)Build the Model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

## 12)Train the Model

```python
lr.fit(x_train,y_train)
```

## 13)Test the Model

```python
y_pred=lr.predict(x_test)
```
